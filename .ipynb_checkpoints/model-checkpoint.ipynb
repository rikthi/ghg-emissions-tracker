```python
import pandas as pd
```

```python
df = pd.read_csv('emissions.csv')
df
```

```python
y = df['GHG Emissions']
y
```

```python
x = df.drop(['GHG Emissions','State','Year'], axis = 1)
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 100)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# Linear Regression

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train_scaled, y_train) 

x_train_scaled
```

```python
y_lr_train_pred = lr.predict(x_train_scaled)
y_lr_test_pred = lr.predict(x_test_scaled)
```

```python
y_lr_train_pred
```

```python
from sklearn.metrics import r2_score

lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_r2 = r2_score(y_test, y_lr_test_pred)

print(f'train r2 = {lr_train_r2}\ntest r2 = {lr_test_r2}')
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2, random_state=100)

rf.fit(x_train_scaled, y_train)
```

```python
y_rf_train_pred = rf.predict(x_train_scaled)
y_rf_test_pred = rf.predict(x_test_scaled)
```

```python
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_r2 = r2_score(y_test, y_rf_test_pred)

print(f'train r2 = {rf_train_r2}\ntest r2 = {rf_test_r2}')
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.scatter(x=y_train, y=y_lr_train_pred,  alpha=0.5)

plt.plot()

plt.ylabel('Predicted GHG Emissions LR')
plt.xlabel('Actual GHG Emissions')
```

```python

plt.scatter(x=y_train, y=y_rf_train_pred,  alpha=0.5)

plt.plot()

plt.ylabel('Predicted GHG Emissions RF')
plt.xlabel('Actual GHG Emissions')
```
